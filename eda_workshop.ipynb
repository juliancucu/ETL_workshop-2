```python
import pandas as pd

# Cargar los datasets
spotify_df = pd.read_csv('spotify_dataset.csv')
grammys_df = pd.read_csv('the_grammy_awards.csv')

# Exploración inicial: Spotify
print("--- Spotify Dataset ---")
print(spotify_df.info()) # Revisa tipos de datos y valores no nulos
print("\nValores nulos por columna:\n", spotify_df.isnull().sum())
print("\nFilas duplicadas:", spotify_df.duplicated().sum())
display(spotify_df.head())

# Exploración inicial: Grammys
print("\n--- Grammys Dataset ---")
print(grammys_df.info()) # Revisa tipos de datos y valores no nulos
print("\nValores nulos por columna:\n", grammys_df.isnull().sum())
print("\nFilas duplicadas:", grammys_df.duplicated().sum())
display(grammys_df.head())
```

```python
import sqlite3
import pandas as pd

# Volvemos a leer el archivo CSV de los Grammys
grammys_df = pd.read_csv('the_grammy_awards.csv')

# Creamos una conexión a nuestra nueva base de datos local
# (Si el archivo 'grammys.db' no existe, Python lo creará automáticamente)
conn = sqlite3.connect('grammys.db')

# Cargamos el DataFrame completo dentro de la base de datos en una tabla
# llamada 'grammys_table'
grammys_df.to_sql('grammys_table', conn, if_exists='replace', index=False)

# Verificamos leyendo desde la base de datos para asegurarnos de que funcionó
test_query = pd.read_sql_query("SELECT * FROM grammys_table LIMIT 5", conn)
print("¡Lectura exitosa desde la base de datos!")
display(test_query)

# Cerramos la conexión
conn.close()
```

```python
# LIMPIEZA DE SPOTIFY 
spotify_df = spotify_df.drop(columns=['Unnamed: 0'], errors='ignore')

# Eliminar nulos SOLO en las llaves de cruce (es solo 1 fila de 114,000)
spotify_df = spotify_df.dropna(subset=['artists', 'track_name'])

# Imputar nulos en columnas descriptivas
spotify_df['album_name'] = spotify_df['album_name'].fillna('Álbum Desconocido')

# Estandarizar
spotify_df['artists_clean'] = spotify_df['artists'].str.lower().str.strip()
spotify_df['track_name_clean'] = spotify_df['track_name'].str.lower().str.strip()

# LIMPIEZA DE GRAMMYS
# Eliminar nulos SOLO en las llaves de cruce
grammys_df = grammys_df.dropna(subset=['artist', 'nominee'])

# Imputar nulos en columnas descriptivas (Aquí salvamos muchos datos)
grammys_df['workers'] = grammys_df['workers'].fillna('Sin registro de trabajadores')
grammys_df['img'] = grammys_df['img'].fillna('Sin imagen')

# Estandarizar
grammys_df['artist_clean'] = grammys_df['artist'].str.lower().str.strip()
grammys_df['nominee_clean'] = grammys_df['nominee'].str.lower().str.strip()

# FUSIÓN DE DATOS (MERGE) 
merged_df = pd.merge(
    spotify_df, 
    grammys_df, 
    left_on=['artists_clean', 'track_name_clean'], 
    right_on=['artist_clean', 'nominee_clean'], 
    how='inner' 
)

# Limpieza final
merged_df = merged_df.drop(columns=['artists_clean', 'track_name_clean', 'artist_clean', 'nominee_clean'])

print(f"¡Fusión exitosa con limpieza profesional! Filas: {merged_df.shape[0]}, Columnas: {merged_df.shape[1]}.")
display(merged_df.head(3))
```

```python
import sqlite3

#GUARDAR PARA GOOGLE DRIVE
# Guardamos el dataframe completo como CSV
merged_df.to_csv('final_dataset_spotify_grammys.csv', index=False)
print("Archivo CSV guardado con éxito para Google Drive.")

# DISEÑO DEL MODELO DE ESTRELLA (STAR SCHEMA) 

# Dimensión 1: Canciones (Tracks)
dim_track = merged_df[['track_id', 'track_name', 'album_name', 'track_genre', 'duration_ms', 'explicit']].drop_duplicates()

# Dimensión 2: Artistas (Artists) - Creamos un ID numérico
dim_artist = merged_df[['artists']].drop_duplicates().reset_index(drop=True)
dim_artist['artist_id'] = dim_artist.index + 1

# Dimensión 3: Premios (Awards) - Creamos un ID numérico
dim_award = merged_df[['title', 'year', 'category', 'winner']].drop_duplicates().reset_index(drop=True)
dim_award['award_id'] = dim_award.index + 1

# Tabla de Hechos (Fact Table): Cruzamos para traer los IDs y dejamos solo los números
fact_metrics = merged_df[['track_id', 'artists', 'title', 'year', 'popularity', 'danceability', 'energy', 'tempo']]
fact_metrics = pd.merge(fact_metrics, dim_artist, on='artists', how='left')
fact_metrics = pd.merge(fact_metrics, dim_award, on=['title', 'year'], how='left')

# Dejamos solo las llaves (IDs) y las métricas
fact_table = fact_metrics[['track_id', 'artist_id', 'award_id', 'popularity', 'danceability', 'energy', 'tempo']]

# CARGAR AL DATA WAREHOUSE (SQLite) 
# Creamos una nueva base de datos que actuará como nuestro Data Warehouse
dw_conn = sqlite3.connect('data_warehouse.db')

# Subimos todas las tablas
dim_track.to_sql('dim_track', dw_conn, if_exists='replace', index=False)
dim_artist.to_sql('dim_artist', dw_conn, if_exists='replace', index=False)
dim_award.to_sql('dim_award', dw_conn, if_exists='replace', index=False)
fact_table.to_sql('fact_performance', dw_conn, if_exists='replace', index=False)

dw_conn.close()
print("Data Warehouse construido exitosamente con Modelo de Estrella.")
```
